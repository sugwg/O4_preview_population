```python
#import logging, sys
#logging.disable(sys.maxsize)

from pycbc import init_logging
init_logging(False)
from gworacle import query, mass, snr
from astropy.utils.data import download_file
import json, numpy


data = query.get_gracedb_superevents(cache=True)

# Downselect event gracedb
far_threshold = 1e-7 # This will exclude a lot of real BBH signals, but only the quiet ones

prod = [k for k in data if k['category'] == 'Production' and ('DQOK' in k['labels'] or 'ADVOK' in k['labels'])]
print("Production events", len(prod))

sig = [k for k in prod if k['far'] <= far_threshold]
print("Pass FAR threshold", len(sig))

ret = []
for s in sig:
    root_url = s['links']['files']
    fdir = download_file(root_url, cache=True)
    flist = json.load(open(fdir, 'r'))
    if 'Retraction' in ' '.join(flist):
        ret.append(s)

# Remove the retracted ones
for s in ret:
    sig.remove(s)
print("After removing retractions", len(sig))

# Organize by id
events = {k['superevent_id']: {'data':k} for k in sig}
```

```python
import logging, sys
logging.disable(sys.maxsize)
import tqdm
import numpy
from matplotlib import pyplot as plt
from scipy.interpolate import interp1d
from scipy.stats import norm

from gworacle.range import get_ranges
from gworacle.psd import extract_psd_from_image
from gworacle.query import get_gwosc_spectrum_url
from gworacle.sky import get_gracedb_skymap
import gworacle.snr

from astropy.utils.data import download_file
from astropy.table import QTable
import astropy_healpix

from pycbc.detector import Detector
from pycbc.waveform import get_fd_waveform
from pycbc.filter import sigma
from pycbc.cosmology import redshift
from pycbc.waveform.spa_tmplt import spa_distance

# pregenerate some reference templates
m = numpy.geomspace(1, 575, 300)
waveforms = {}
waveforms_hs = {}
inc = 0.64

for mass in m:
    waveforms[(inc, mass)] = get_fd_waveform(approximant="IMRPhenomD", mass1=mass, mass2=mass,
                             f_lower=20, delta_f=1.0/8, 
                             inclination=inc,
                             distance=1)

for mass in m:
    waveforms_hs[(inc, mass)] = get_fd_waveform(approximant="IMRPhenomD", mass1=mass, mass2=mass,
                             f_lower=20, delta_f=1.0/8, spin1z=0.5, spinz2=0.5,
                             inclination=inc,
                             distance=1)

rdata = {}
for i, event in tqdm.tqdm(enumerate(events), total=len(events.keys())):
    time = events[event]['data']['t_0']
    ifos = events[event]['data']['preferred_event_data']['instruments'].split(',')

    # Get psds for this time # special cases because no info present
    offset = 0
    if event == 'S230919bj':
        offset = -86400
    
    try:
        fname = download_file(get_gwosc_spectrum_url(time + offset), cache=True)
    except: # Some images are missing, if so try the next day
        fname = download_file(get_gwosc_spectrum_url(time + 86400 + offset), cache=True)

    # Get the psds
    psds = extract_psd_from_image(fname, target_f_min=10, target_f_max=4000, delta_f=1.0/8)

    try:
        ranges = get_ranges(time)
    except:
        print("FAILED TO GET RANGE", event, time, ifos)
        ranges = None
    
    # Sanity check that the psds make sense, else break immediately.
    failed_psd = False
    for ifo in ifos:
        d = spa_distance(psds[ifo], 1.4, 1.4, 20.0)
        if d > 500 or d < 50:
            print(event, ifo, d, "PSD RANGE DOESN'T MAKE SENSE, check extraction!!!!, skipping")
            failed_psd = True
 
        if ranges is not None:
            # adjust psd range ot match inspiral, psd can easily be a pixel or two off which matters in log scale
            fac = (d / 2.26) / ranges[ifo]
            if abs(1 - fac) > 0.3:
                print(event, d / 2.26, ranges[ifo], fac, "WOAH, trying to rescale PSD by a large amount!!!")
            else:
                psds[ifo] = psds[ifo] * fac * fac
    
    if failed_psd:
        continue


    # Get the skymap

    # try multiple and then compare
    skymaps = {}
    for type in ['bayestar', 'PublicationSamples', 'bilby', 'lalinference']:
        try:
            mname = get_gracedb_skymap(event, map_type=type, use_cache=True)
            skymap = QTable.read(mname)

            if 'PROBDENSITY' in skymap.dtype.names:
                level, ipix = astropy_healpix.uniq_to_level_ipix(skymap['UNIQ'])
                nside = astropy_healpix.level_to_nside(level)
                ra, dec = astropy_healpix.healpix_to_lonlat(ipix, nside, order='nested')
                pixel_area = astropy_healpix.nside_to_pixel_area(nside)
                probd = skymap['PROBDENSITY']
            elif 'PROB' in skymap.dtype.names:
                nside = int((len(skymap['PROB'])/ 12)**(1/2))
                ra, dec = astropy_healpix.healpix_to_lonlat(numpy.arange(0, len(skymap['PROB'])), nside)
                pixel_area = astropy_healpix.nside_to_pixel_area(nside)
                probd = skymap['PROB'] / pixel_area

            pix = probd.argmax() 
            dr = .1
            r = numpy.arange(0, 16000, dr)
            d = r**2.0  * skymap['DISTNORM'][pix] * norm(skymap['DISTMU'][pix],
                                                         skymap['DISTSIGMA'][pix]).pdf(r) * dr
            dist = r[d.argmax()]

            skymaps[type] = skymap, dist, ra, dec, pix, probd, pixel_area
        except Exception as e:
            pass

    skymap_flipped = False
    if "bayestar" in skymaps:
        skymap, dist, ra, dec, pix, probd, area = skymaps['bayestar']
        good_map = True
        for type in ['PublicationSamples', 'bilby', 'lalinference']:
            if type in skymaps:
                _, dist2, _, _, _, _, _ = skymaps[type]

                if (dist2 - dist) / dist2 > 0.3: 
                    print(event, dist, dist2, "Significant difference from bayestar and ", type)
                    skymap, dist, ra, dec, pix, probd, area = skymaps[type]
                    skymap_flipped = True
                    break
    print(len(area), len(ra))                 
    # get snr estimate
    st = gworacle.snr.get_bayestar_snr(events[event]['data'])   
    if st['snr'] is None:
        print(event, "failed to find SNR information, skipping")
        continue
    
    # Basic mass estimate
    pol = 0
    mest = []
    fp = {}
    fc = {}
    for ifo in ifos:
        fp[ifo], fc[ifo] = Detector(ifo).antenna_pattern(ra[pix].value, dec[pix].value, 0, time)

    rdata[event] = {'dist':dist, 
                    'distmu': skymap['DISTMU'][pix].value, 
                    'distsigma': skymap['DISTSIGMA'][pix].value,
                    'distnorm': skymap['DISTNORM'][pix].value,
                    'snr': st, 'time': time, 'skyflip': skymap_flipped,
                     'far': events[event]['data']['far'],
                     'ifos':ifos}
    
    for j, wf in enumerate([waveforms, waveforms_hs]):
        mest = []
        starg = st['snr']
        nsnrs = []
        for mass in m:
            hp, hc = wf[(inc, mass)]
            hp = hp / dist
            hc = hc / dist
        
            nsnrsq = 0
            for ifo in ifos:      
                h = fp[ifo] * hp + fc[ifo] * hc
                snr = sigma(h, psd=psds[ifo], low_frequency_cutoff=20, high_frequency_cutoff=1000)
                nsnrsq += snr**2.0    
            nsnr = nsnrsq ** 0.5
            nsnrs.append(nsnr)
        
        nsnrs = numpy.array(nsnrs)


        from matplotlib import pyplot as plt
        from pycbc.cosmology import redshift
        plt.plot(m, nsnrs)
        print(event, starg, dist, redshift(dist))
    
        ipeak = nsnrs.argmax()
        mpeak = m[ipeak]
        speak = nsnrs[ipeak]
        mlimit = 500
        
        f = interp1d(nsnrs[:ipeak], m[:ipeak], fill_value=(m.min(), mpeak), bounds_error=False)
        mest.append(float(f(starg)))
    
        der = (nsnrs[1:] - nsnrs[:-1]) / (m[1:] - m[:-1])
        mp = numpy.where(der < 0.1 / 3)[0]


        if len(mp) != 0:
            mlimit = m[mp[0]]
            if m[mp[0]] < f(starg):
                print(event, "MASS LIMIT MAYBE", mlimit, f(starg), starg, dist)

        sfail = False
        if speak < starg:
            sfail = True
            print(event, "SNR not achieved, assumptions about source are likely wrong")

        if j == 0:
            rdata[event]['mest'] = mest
            rdata[event]['mlimit'] = mlimit
            rdata[event]['sfail'] = sfail
        if j == 1:
            rdata[event]['mest2'] = mest
            rdata[event]['mlimit2'] = mlimit
            rdata[event]['sfail2'] = sfail
```

```python
import json
json.dump(rdata, open('massest.json', 'w'))
```
